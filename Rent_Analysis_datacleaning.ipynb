```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('final_rentanalysis_project11.csv')
```

```python
df.head()
```

```python
# bhk
df['BHK'] = df['Property_title'].str.extract(r'(\d+)\s*BHK', expand=False)
df['BHK'] = df['BHK'].astype('Int64')
```

```python
df['rent'].unique()
```

```python
def clean_rent(value):
    if pd.isna(value):
        return None

    value = value.replace('₹', '').replace(',', '').strip().lower()

    if 'lac' in value:
        num = float(value.replace('lac', '').strip())
        return num * 100000   # convert lac to rupees

    else:
        return float(value)
```

```python
df['rent_numeric'] = df['rent'].apply(clean_rent)
```

```python
df[['rent', 'rent_numeric']].sample(10)
```

```python
df["Property_title"].unique()
```

```python
df['property_type'] = df['Property_title'].str.extract(r'BHK\s+(\w+)')
```

```python
df['property_type']
```

```python
df['locality'] = (
    df['Property_title']
    .str.split(' in ')
    .str[-1]
    .str.replace(', Pune', '', regex=False)
)
```

```python
df['locality'].unique()
```

```python
df['locality'] = (
    df['locality']
    .str.split(',')
    .str[-1]
    .str.strip()
)
```

```python
df['locality']
```

```python
df['availability_type'] = df['availability'].replace({
    'Immediately': 'Immediate',
    "From Feb '26": 'Future'
})
```
