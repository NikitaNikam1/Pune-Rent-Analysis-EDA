```python
url = 'https://www.magicbricks.com/property-for-rent/residential-real-estate?&proptype=Multistorey-Apartment,Builder-Floor-Apartment&cityName=Pune&showPrimePropsinFixedSlotsSEO=N&mbTrackSrc=tabChange&isNRI=N&intent-text=flats&page=1&category=R&multiLang=en'
for i in range(1,116):
    print(f'https://www.magicbricks.com/property-for-rent/residential-real-estate?&proptype=Multistorey-Apartment,Builder-Floor-Apartment&cityName=Pune&showPrimePropsinFixedSlotsSEO=N&mbTrackSrc=tabChange&isNRI=N&intent-text=flats&page=1&category=R&multiLang=en={i}')
```

```python
title_data = []
rent_data = []
furnishing_data = []
tenant_data = []
carpet_area_data = []
facing_data = []
balcony_data = []
availability_data = []
bathroom_data = []
floor_data = []
point_of_contact = []


for i in range(1, 100):
    url = f'https://www.magicbricks.com/property-for-rent/residential-real-estate?&proptype=Multistorey-Apartment,Builder-Floor-Apartment&cityName=Pune&showPrimePropsinFixedSlotsSEO=N&mbTrackSrc=tabChange&isNRI=N&intent-text=flats&page=1&category=R&multiLang=en={i}'
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
    }

    response = requests.get(url, headers=headers)
    if response.status_code == 403:
        print("403 Error")
        break

    soup = BeautifulSoup(response.text, "html.parser")
    containers = soup.find_all('div', class_='mb-srp__list')

    for c in containers:

        # ----------------------- TITLE ------------------------------
        property_title = c.find('h2', class_='mb-srp__card--title')
        title_data.append(property_title.text.strip() if property_title else np.nan)

        # ---------------------- RENT -------------------------------
        rent = c.find('div', class_='mb-srp__card__price--amount')
        rent_data.append(rent.text.strip() if rent else np.nan)


        # -------------------- CARPET AREA -----------------------------
        parent = c.find("div", {'data-summary': 'carpet-area'})
        if parent:
            carpet_area = parent.find("div", class_="mb-srp__card__summary--value")
            carpet_area_data.append(carpet_area.text.strip() if carpet_area else np.nan)
        else:
            carpet_area_data.append(np.nan)

        # ------------------- FURNISHING ----------------------------
        parent = c.find("div", {'data-summary': 'furnishing'})
        if parent:
            furnishing = parent.find("div", class_="mb-srp__card__summary--value")
            furnishing_data.append(furnishing.text.strip() if furnishing else np.nan)
        else:
            furnishing_data.append(np.nan)

        # ---------------------- BALCONY ---------------------------------
        parent = c.find("div", {'data-summary': 'balcony'})
        if parent:
            balcony = parent.find("div", class_="mb-srp__card__summary--value")
            balcony_data.append(balcony.text.strip() if balcony else np.nan)
        else:
            balcony_data.append(np.nan)    
            
        # ------------------ TENANT PREFFERED ------------------------
        parent = c.find("div", {'data-summary': 'tenent-preffered'})
        if parent:
            tenant_preffered = parent.find("div", class_="mb-srp__card__summary--value")
            tenant_data.append(tenant_preffered.text.strip() if tenant_preffered else np.nan)
        else:
            tenant_data.append(np.nan)


        # --------------------- FACING ----------------------------------
        parent = c.find("div", {'data-summary': 'facing'})
        if parent:
            facing = parent.find("div", class_="mb-srp__card__summary--value")
            facing_data.append(facing.text.strip() if facing else np.nan)
        else:
            facing_data.append(np.nan)

        
        # --------------------- AVAILABILITY -----------------------------
        parent = c.find("div", {'data-summary': 'status'})
        if parent:
            availability = parent.find("div", class_="mb-srp__card__summary--value")
            availability_data.append(availability.text.strip() if availability else np.nan)
        else:
            availability_data.append(np.nan)    

        # -------------------- Bathroom -----------------------------------
        parent = c.find("div", {'data-summary': 'bathroom'})
        if parent:
            bathroom = parent.find("div", class_="mb-srp__card__summary--value")
            bathroom_data.append(bathroom.text.strip() if bathroom else np.nan)
        else:
            bathroom_data.append(np.nan)


        # ------------------- Floor ----------------------------------------
        parent = c.find("div", {'data-summary': 'floor'})
        if parent:
            floor = parent.find("div", class_="mb-srp__card__summary--value")
            floor_data.append(floor.text.strip() if floor else np.nan)
        else:
            floor_data.append(np.nan)

        # -------------------- point_of_Contact -----------------------------   
        contact = c.find('span', class_='mb-srp__action--btn medium btn-red')
        point_of_contact.append(contact.text.strip() if contact else np.nan)
```

```python
print("property_title",len(title_data))
print("rent",len(rent_data))
print("furnishing",len(furnishing_data))
print("carpet_area",len(carpet_area_data))
print("facing",len(facing_data))
print("balcony",len(balcony_data))
print("availability",len(availability_data))
print("batrooms",len(bathroom_data))
print("floor",len(floor_data))
print("point_of_contact",len(point_of_contact))
```

```python
df = pd.DataFrame({"Property_title": title_data,
                   'rent': rent_data,
                   'furnishing': furnishing_data,
                   'carpet_area': carpet_area_data,
                   'facing': facing_data,
                   'balcony': balcony_data,
                   'availability':availability_data,
                   'bathrooms':bathroom_data,
                   'floors':floor_data,
                   'point_of_contact': point_of_contact})
```
